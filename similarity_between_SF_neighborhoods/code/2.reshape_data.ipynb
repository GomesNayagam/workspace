# DATA RESHAPING

In this notebook, I am going to reshape the datasets and prep them for a merge in the next document.  Some of the datasets have information over time.  I want some datasets to have totals (regardless of time of occurence), such as the total # of businesses that are active in a hood, regardless when they opened.  And in other data sets, I want to keep the time information in.  For example, how many businesses of a particular type opened in 2013.  

I also want to bin the time data by year, because, in this first pass at least, I am going to look at courser effects of time. 


The following are the datasets I will look at this notebook:

1) restaurant (a. restaurant_total & b. restaurant_time);
2) crime (crime_time);
3) quality (quality_time);
4) housing (a. house_price_time & b. freq_price_time);
5) business (a. biz_time & b. biz_total).

# RESTAURANTS 

From the biz document, I am going to make two datasets related to # of restaurants per hood.  Total number open and total opening per year. 

```python
#make a list of just the restaurants
restaurantcodes=['3079', '5800', '5812']
restaurants=biz[biz['PBC Code'].isin(restaurantcodes)]
restaurants.head(n=2)
```

```python
#CREATE A LIST OF THE TOTAL NUMBER OF RESTAURANTS THAT ARE ACTIVE IN EACH HOOD, REGARDLESS OF START DATE
restaurants_tot_prep1 = restaurants.groupby(['hood']).count()
restaurant_total = restaurants_tot_prep1[[0]]
restaurant_total.rename(columns={'Ownership_Name' : 'restaurant_total'}, inplace=True)

#create a for each class type on each date
rest_prep = restaurants.groupby(['hood', 'Date']).count()
rest_prep1=rest_prep[[0]]

#rename column
rest_prep1.rename(columns={'Ownership_Name' : 'restaurant_time'}, inplace=True)
#reset index
rest_prep1=rest_prep1.reset_index()
#make only date the index
rest_date=rest_prep1.set_index('Date')
#make it type datetime, not string
rest_date.index = pd.to_datetime(rest_date.index)

#get ride of dates before 2007 cuz none of my other data go past that point
rest_recent_dates=rest_date.ix['2007-07-01':'2013-12-01']
#rest_recent_dates.head()

#bin years starting on the last calender day of june - giving me 5 years of data for each hood (117 hoods)
rest_year1=rest_recent_dates.groupby(['hood']).resample('A-JUN', how='sum').fillna(0)

#need to get rid of year that ends on june-2014, cuz it doesn't have enought data.
rest_year2=rest_year1.reset_index(0)
restaurant_time=rest_year2.ix['2007':'2013']

#make csv copies of restaurant data
restaurant_time.to_csv('../hoods_data/prep_to_analyze/restaurant_time.csv')
restaurant_total.to_csv('../hoods_data/prep_to_analyze/restaurant_total.csv')

restaurant_time.head()
```

# CRIME

I am going to get counts of crime grouped by hood, date, and then category. 

```python
#create a for each class type on each date
crime_unstacked = crime.groupby(['hood', 'Date', 'Category']).count()

crime_unstacked1=crime_unstacked[[0]]

#if missing data, assume no crime occurred.  Fill in Nan's with 0.
crime_unstacked2=crime_unstacked1.unstack().fillna(0)
#collapse hierachical column index
crime_unstacked2.columns = crime_unstacked2.columns.get_level_values(1)
crime_unstacked2.reset_index(inplace=True)

#I want to bin the data together by year
crime_date = crime_unstacked2[1:]
crime_date.set_index('Date', inplace=True)
crime_date.index = pd.to_datetime(crime_date.index)
crime_time=crime_date.groupby(['hood']).resample('A-JUN', how='sum').fillna(0)
crime_time.reset_index(inplace=True)
crime_time.rename(columns={'level_1' : 'Date'}, inplace=True)
```

Other works has shown that crime can be reduced into the following categories.  So I will collapse across these five factors.

```python
crime_dict ={'violent': ["ASSAULT", "ROBBERY", 'KIDNAPPING', 'SEX OFFENSES, FORCIBLE', 'EXTORTION'], 
             'theft':['ARSON','LARCENY/THEFT', 'VEHICLE THEFT', 'BURGLARY','VANDALISM','STOLEN PROPERTY'],
             'disorderly':['DISORDERLY CONDUCT','DRUG/NARCOTIC','PROSTITUTION','DRUNKENNESS','LIQUOR LAWS', \
                         'LOITERING','SEX OFFENSES, NON FORCIBLE','GAMBLING','PORNOGRAPHY/OBSCENE MAT'],
             'hedonistic':['ARSON','LARCENY/THEFT', 'VEHICLE THEFT', 'BURGLARY','STOLEN PROPERTY'],
             'money_fraud':['BRIBERY','BAD CHECKS', 'FRAUD',	'EMBEZZLEMENT',	'FORGERY/COUNTERFEITING']}


for k,v in crime_dict.items():
    crime_time[k]=crime_time[v].sum(axis=1)

crime_time = crime_time[['Date', 'hood', 'violent', 'disorderly', 'money_fraud', 'hedonistic', 'theft']]

#write to csv
crime_time.to_csv('../hoods_data/prep_to_analyze/crime_time_full.csv')
crime_time.ix[:2, :20]
```

# QUALITY

I am going to transfore and categorize the 311 complaints by hoods and years

```python
#create a for each class type on each date
qual_prep = quality.groupby(['hood', 'Date', 'Category']).count()
qual_prep1=qual_prep[[0]]

qual_prep1.rename(columns={'Date' : 'qualitycomplaints_time'}, inplace=True)

#convert back to wide format
quality_unstacked2=qual_prep1.unstack().fillna(0)

#remove the hierarchical index of the column names
levels = quality_unstacked2.columns.levels
labels = quality_unstacked2.columns.labels
quality_unstacked2.columns = levels[1][labels[1]]

#reset the index
quality_prep1=quality_unstacked2.reset_index()
#so that I can, now make date the only index
quality_date=quality_prep1.set_index('Date')
#make it type datetime, not string
quality_date.index = pd.to_datetime(quality_date.index)

#bin years starting on the last calender day of june - giving me 5 years of data for each hood (117 hoods)
quality_time_full=quality_date.groupby(['hood']).resample('A-JUN', how='sum').fillna(0)
quality_time_full=quality_time_full.reset_index(0)


quality_time_full.to_csv('../hoods_data/prep_to_analyze/quality_time_full.csv')
quality_time_full.ix[:5, :20]
```
