I know from my previous analyses done on my similarity matrix that the features I have created are very good at making sensible predictions about the similarities between hoods in SF (sensibilty is assessed by my domain knowledge of SF hoods).  What kind of insights can I mine out of these quantitative measures of neighborhood similarities?


### Which hoods, when compared to their top-5 most similar hoods, are abnormally cheap & expensive?

What if we compared the value of neighborhood to its top 5 most similar hoods - its peers? Would it be cheaper or more expensive to its peers? What if we harnessed the power of my similarity matrix and went through every hood in the city, comparing the value of each hood to the average value of its top-5 most similar hoods? If we do this, we obtain a value - the price difference for each hood, which can then be converted into z-scores to see if any of them are abnormally cheap or expensive (e.g., zscore greater than 2.5). These hoods may be under or overvalued.

```python
#open up the similarity matrix calculated in previous ipython notebook
sf_hoods=pd.read_csv(open('/Users/anson/Documents/Office/zipfian/repos/Personal_project/hoods_data/output/hoods_similarity_renamed.csv'))
sf_hoods=sf_hoods.set_index(['hoods'])
```

```python
#function to find price differences between target and similar hoods
def peer_evaluation(housing_data, column_name, comparison_number):
    prices_dict = {}
    error =[]
    #go through each hood in the matrix (111 total)
    for (target_name, col) in sf_hoods.iteritems():
       
        #the housing data is missing some hoods, so use len to check to see if hood is there (1 = yes; 0 = no)
        if len(housing_data[housing_data['hood'] == target_name]) == 1:
            #get the logged value of the target hood
            target_prices= housing_data[housing_data['hood'] == target_name]
            target_price = math.log(target_prices[column_name])
       
            #get a list of the most similar hoods (their indexes), using comparison number to determine how many to return 
            #by sorting a row of the similarity matrix;  this return a pd series with hood names as the index
            sim_hoods = np.sort(sf_hoods[target_name])[1:comparison_number+1]
            
                
            #now that I have the names of the similar hoods (sim_hoods.index),
            # get house prices on hoods that are sim_hoods.index
            sim_hood_prices = housing_data[housing_data['hood'].isin(sim_hoods.index)]
            
            #get mean of those values;  
            #number in bracks is for column to analyze, eg. # of bedrooms (1-3) in bed_prices file, 
                #or price_sqft in the housing_averages file. 
            sim_hoods_mean_price = np.mean([log(i) for i in sim_hood_prices[column_name] if i > 0])
            
            price_diff = target_price - sim_hoods_mean_price
            prices_dict[target_name] = np.array(price_diff)
        
        elif len(housing_data[housing_data['hood'] == target_name]) == 0:
            error.append('a')
    print 'number of hoods analyzed: '+ str(len(sf_hoods)-len(error))
    final = pd.DataFrame(prices_dict.items())
    final.rename(columns={0: 'hood', 1:'price_diff'}, inplace=True)
    return final
```

Now find calculate z scores to find out which hoods are abnormally cheap and expensive compared to their peers.  

```python
#This function will return outliers; a pandas df contaiing only names of hoods with z-scores over a certain value.
def abnormal_hoods(pandas_df, z_threshold):
    pandas_df['price_diff'] = pandas_df['price_diff'].astype(float)
    mean = pandas_df['price_diff'].mean()
    stdev = pandas_df['price_diff'].std()
    #add zscore column
    pandas_df['zscore'] = pandas_df['price_diff'].apply(lambda x: (x - mean) / stdev)
    #make df containing only abnormal hoods
    abnormal = pandas_df[abs(pandas_df['zscore']) > z_threshold]
    abnormal.sort('zscore', inplace=True, ascending=False)
    return abnormal


Z_THRESHOLD=2
print 'ONE BEDROOM'
print abnormal_hoods(prices_one_bedroom, Z_THRESHOLD)
print ''
print 'TWO BEDROOMS'
print abnormal_hoods(prices_two_bedroom, Z_THRESHOLD)
print ''
print 'THREE BEDROOMS'
print abnormal_hoods(prices_three_bedroom, Z_THRESHOLD)
```
